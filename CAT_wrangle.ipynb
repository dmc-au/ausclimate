```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
demand = pd.read_csv(
    "./Data/ClimateActionTracker/CAT_energy_demand_GJ_cap.csv",
    encoding="ISO-8859-1"
)
demand.columns = ['Year', 'AU', 'World']
```

```python
plt.figure(figsize=(10,10))
plt.plot(demand["Year"], demand["AU"], label='AU')
plt.plot(demand["Year"], demand["World"], label='World')
plt.title('Energy Demand (GJ) Per Capita')
plt.xticks(demand["Year"], rotation=90)
plt.legend()
```
