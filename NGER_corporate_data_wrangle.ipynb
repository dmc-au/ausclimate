```python
dfs = [
    df2021,df2020,df2019,
    df2018,df2017,df2016,
    df2015,df2014 #df2013 is special
]
```

```python
# Dataframe of corporate emissions 2013-2021
years = [
    '2021','2020','2019',
    '2018','2017','2016',
    '2015','2014','2013'
]
s1 = [pd.to_numeric(pd.DataFrame(df).iloc[:,2], errors='coerce').sum() for df in dfs]
s2 = [pd.to_numeric(pd.DataFrame(df).iloc[:,3], errors='coerce').sum() for df in dfs]
total_energy = [pd.to_numeric(pd.DataFrame(df).iloc[:,3], errors='coerce').sum() for df in dfs]

s1.append(pd.to_numeric(pd.DataFrame(df2013).iloc[:,1], errors='coerce').sum())
s2.append(pd.to_numeric(pd.DataFrame(df2013).iloc[:,2], errors='coerce').sum())
total_energy.append(pd.to_numeric(pd.DataFrame(df2013).iloc[:,3], errors='coerce').sum())

df_total = pd.DataFrame(data={
    'years': years,
    'scope 1 sum': s1,
    'scope 2 sum': s2,
    'total energy consumed': total_energy
}).loc[::-1]
df_total
```

```python
# Total scope 1 & 2 CO2
total_scope_sum = df_total['scope 1 sum'] + df_total['scope 2 sum']
total_scope_sum
plt.clf()
plt.bar(df_total['years'], total_scope_sum)
```

```python
# Total energy consumed
plt.clf()
plt.bar(df_total['years'], df_total['total energy consumed'])
```

The reporting for total energy must have changed from 2012-13 to 2013-14

```python
# Stackplot of scope 1 and 2 CO2
plt.clf()
plt.stackplot(df_total['years'], df_total['scope 2 sum'], df_total['scope 1 sum'])
plt.title('Corporate Total Emissions (t CO2-e)')
plt.legend()
```
