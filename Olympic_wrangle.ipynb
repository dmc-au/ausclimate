```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(
    "./Data/Olympic/Kaggle2/athlete_events.csv",
    encoding="ISO-8859-1"
)
```

```python
df
```

```python
country_sum = pd.DataFrame(
    df.groupby(by=['NOC'])['Medal Val'].sum()
).sort_values(by=['Medal Val'], ascending=False)[0:10]
country_sum
```

```python
plt.figure()
country_sum.plot.bar(
    figsize=(10,10),
)
plt.title(
    "Total Olympic Medal Count by Country (1896-2016)",
    size=16
)
plt.xlabel('Country')
# plt.ylabel('Medal count')
plt.legend(
    labels=['Medal count'],
    loc=1,
    prop={'size': 12}
)
plt.show()
```
